```python
import pandas as pd
import geopandas as gp
import numpy as np
import matplotlib.pyplot as plt
```

```python
midterm1 = pd.read_csv('Alldata.csv')
```

```python
midterm1
```

```python
midterm1['lon'] =[-122.33, -122.60, -118.23, -122.92, -104.85, -111.92, -112.12, -87.73, -90.24, -96.73, -70.97, -77.01, -80.22, -68.78 ]
midterm1['lat'] =[46.61, 37.77, 34.05, 45.63, 39.76, 40.77, 33.60, 41.83, 38.65, 32.82, 42.31, 38.89, 25.78, 44.82 ]
```

```python
midterm1
```

```python
cities = gp.GeoDataFrame(midterm1, geometry = gp.points_from_xy(midterm1.lon, midterm1.lat))
cities.geometry
```

```python
filepath = 'maps/states.shp'
states = gp.read_file(filepath)
```

```python
states.crs
```

```python
states = states.to_crs("epsg:4326")
```

```python
fig, ax = plt.subplots(figsize=(20,20))
states.plot(ax=ax,color="lightgray", edgecolor="grey", linewidth=0.4, legend=True)
cities.plot(column = 'Zone',ax=ax, markersize='Poor_Days', alpha=1.0, categorical=False, legend=True)
ax.axis()
plt.title('Air Quality Index Poor Days: 2020', size=20)
plt.savefig('final_proj/AQI2020.png')
#plt.axis()
#plt.show()
```

```python
fig, ax = plt.subplots(figsize=(20,20))
states.plot(ax=ax,color="lightgray", edgecolor="grey", linewidth=0.4, legend=True)
cities.plot(column = 'Zone',ax=ax, markersize='Poor_Days10', alpha=1.0, categorical=False, legend=True)
ax.axis()
plt.title('Air Quality Index Poor Days: 2010', size=20)
plt.savefig('final_proj/AQI2010.png')
#plt.axis()
#plt.show()
```
