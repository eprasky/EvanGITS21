```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
agerange= np.arange(len(AllData['City']))

plt.bar(agerange,AllData['Poor_Days'],color='blue',label='Poor_Days20',width=.3)
plt.bar(agerange+0.3,AllData['Poor_Days10'], color='red',label='Poor_Days10',width=.3)

plt.title('Total Numer of Poor Days By city')
plt.legend()
plt.xlabel('City')
plt.ylabel('Number of Days')
plt.grid(False)
plt.xticks(ticks = agerange,labels =AllData['City'], rotation= 90)
plt.show()
```

```python
agerange= np.arange(len(AllData['City']))

plt.bar(agerange,AllData['AQIMedian'],color='blue',label='AQIMedian20',width=.3)
plt.bar(agerange+0.3,AllData['AQIMedian10'], color='red',label='AQIMedian10',width=.3)

plt.title('AQIMedian By City')
plt.legend()
plt.xlabel('City')
plt.ylabel('Number of Days')
plt.grid(False)
plt.xticks(ticks = agerange,labels =AllData['City'], rotation= 90)
plt.show()
```

```python
labels = ['central', 'east', 'mountain', 'west']
Poor_Days20 = [62, 5, 125, 171]
Poor_Days10 = [174, 66, 121, 135]

x = np.arange(len(labels)) 
width = 0.35

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, Poor_Days20, width, label='Poor_Days20')
rects2 = ax.bar(x + width/2, Poor_Days10, width, label='Poor_Days10')

ax.set_ylabel('Number of Days')
ax.set_xlabel('Zone')
ax.set_title('Total Number of Poor Days by Zone')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

fig.tight_layout()

plt.show()
```

```python
labels = ['central', 'east', 'mountain', 'west']
AQIMedian20 = [157, 157, 183, 204]
AQIMedian10 = [197, 184, 186, 201]

x = np.arange(len(labels)) 
width = 0.35

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, AQIMedian20, width, label='AQIMedian20')
rects2 = ax.bar(x + width/2, AQIMedian10, width, label='AQIMedian10')

ax.set_ylabel('Number of Days')
ax.set_xlabel('Zone')
ax.set_title('AQIMedian by Zone')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

fig.tight_layout()

plt.show()
```
